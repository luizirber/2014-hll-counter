# HLL notes

## Summary

 - Murmur3 is faster than SHA-1
 - Bias correction works well for SHA-1, not so well for Murmur3.
 - XOR revcomp and forward hashes gives the best results when unifying hashes.

## Next steps

Parallel reading gives linear improvement again, since we can create one HLL per thread, count and merge them later for final result. This scales for as many threads the computer can handle.

# Testing hash functions for HLL

All functions have two variations, one revcomp-aware and another considering only the forward hash.

revcomp version hashes both forward and revcomp and XOR the hashes.
```
   h = hash(kmer)
   r = hash(revcomp)
   return h ^ r
```

Returning max(h, r) messed the counting, still need to figure out why.

````python
import pandas as pd
import numpy as np
from IPython.core.display import Image
import dateutil
````

````python
def parse_output(measures):
    filename = None
    data = {}
    details = {}
    for line in measures.stdout.split('\n'):
        line = line.strip()
        if filename is None:
            filename = line
        elif not line:
            data[filename] = details
            filename = None
            details = {}
        elif 'nique' in line:
            d = line.split(',')
            field, value = d[0].split(':')
            details[field] = int(value)
        elif 'cpu' in line and 'total' in line:
            d = line.split()
            details['time'] = d[-2]
            
    df = pd.DataFrame(data).T
    if "Unique" in df:
        df['Error Cpp'] = abs(df['HLL cpp unique'] - df['Unique']) / df['Unique']
        df['Error Py'] = abs(df['HLL lib unique'] - df['Unique']) / df['Unique']
        

    return df
````
